```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import NullLocator, LinearLocator, MultipleLocator, IndexLocator, FixedLocator, LogLocator,MaxNLocator
```

```python
fig,axs=plt.subplots(2,2,figsize=(15,7))
fig.set_facecolor('yellow')
axs[0,0].grid()
ax00 = axs[0,0]
axs[0,1].set_title('each graph')

ax00.plot(np.arange(0,7),[1,2,3,34,4,67,9])
axs[1,0].plot(np.arange(1,5,0.1))
```

```python
x = np.linspace(0, 10, 100)
y = np.sinc(x) + 0.5
```

```python
fig2 = plt.figure()
fig2.set_facecolor('gray')
fig2.set_size_inches(15, 8)

gs = plt.GridSpec(3,
                  3,
                  figure=fig2,
                  height_ratios=[2, 1, 1],
                  width_ratios=[2, 1, 1])
ax1 = fig2.add_subplot(gs[0, :])
plt.plot(x,x**2)
plt.box(True)
ax2 = plt.subplot(gs[2:3, 0])
ax3 = fig2.add_subplot(gs[1:3, 1:3])

ax1.plot([0,20],[1,10])
ax1.xaxis.set_major_locator(NullLocator())
ax1.yaxis.set_major_locator(LogLocator(base=2))

ax2.scatter(x,
            3 * np.random.random(len(x)) + x * 0.2,
            marker='s',
            label='my data')
ax2.grid()
ax2.legend(bbox_to_anchor=(0.5,0.7))
ax2.xaxis.set_major_locator(LinearLocator(11))
ax2.yaxis.set_major_locator(MultipleLocator(base=0.5))

ax3.set_xlim(0.2, 10)
ax3.set_xlabel('x label')
ax3.plot(x, y, label='Graph')
ax3.scatter(x, (np.random.random(len(x)) - 0.5) * 0.3 + y, c='red',label='Scatter')
ax3.legend(loc='upper left',facecolor='yellow'
    
)
ax3.set(ymargin=0.1)
ax3.set_xlim(xmax=7)
ax3.set_ylim(ymin=0.2)
ax3.xaxis.set_major_locator(IndexLocator(base=0.7,offset=0.3))
ax3.yaxis.set_major_locator(FixedLocator([0,0.1,0.3,0.44,0.6,0.7,1]))
ax3.grid()

ax4=plt.subplot(gs[1,0])
plt.plot(x,x**np.sinc(x))

ax4.xaxis.set_major_locator(MultipleLocator(base=0.2,))
ax4.xaxis.set_major_locator(MaxNLocator(20))
```
