```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(-10,10,0.05)
y=np.sinc(x)
plt.plot(x,y,'r--+',)
plt.fill_between(x,y,-0.1,where=(x>0))
plt.grid()
```

Transfer to artist

```python
plot2 = plt.plot(x,x)
plt.setp(plot2[0],linestyle='-',marker='s')
```

```python
ax1 = plt.subplot(2, 1, 1)
plt.plot(np.random.random(10), '--r')
ax2 = plt.subplot(2, 1, 2)
plt.plot(np.random.random(10), '-b')
plt.subplot(2, 1, 1)
plt.xlim(0, 8)
ax2.grid()
ax2.set_title('Second title')
plt.ylabel('y1 ')
plt.title('First title')
```
