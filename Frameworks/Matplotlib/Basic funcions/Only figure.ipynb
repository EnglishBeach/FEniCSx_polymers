```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-1,5,10)
```

```python
fig= plt.figure()
fig.set(facecolor='#BBB',size_inches=(10,10))

gs=plt.GridSpec(3,3,figure=fig)
ax1= plt.subplot(gs[0,0:2])
ax2=plt.subplot(gs[0,2])
ax3=plt.subplot(gs[1,0])
ax4 = plt.subplot(gs[1, 1])
```

```python
ax1.step(x,2*x,'-go')

ax2.stackplot(x,x**3,x**2,x**2)

ax3.stem(x,-np.sin(x),bottom=0.5,markerfmt='^b',basefmt='-r',linefmt='-g')
ax3.grid()
fig
```

```python
x_rn = np.random.random(20)
y_rn = np.random.random(20)
# ax4.clear()
ax4.scatter(x_rn, y_rn)
fig
```
