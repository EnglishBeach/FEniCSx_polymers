```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(7, 7))
fig.set_facecolor('gray')
ax = plt.subplot()
plt.suptitle('Title!')
ax.set_title('Graph title')
ax.set_ylabel('asd')

plt.figtext(0.3,0.3,'Text of figure',fontsize=24,c='red')
ax.set(ylabel='Oy',xlabel='Ox')

ax.text(0.1,0.6,'Text of axes',fontsize=32,c='yellow',
    bbox=dict(
        boxstyle='darrow',
        alpha=0.5,
        color='green',
        fill=True,
        hatch='*')
    )
ax.annotate(
    'Annotation',
    xy=[0.6, 0.8],
    # xycoords='data',
    xytext=(0.5,0.5),
    fontsize=20,
    # textcoords='offset points',
    arrowprops=dict(
        arrowstyle="->",
        connectionstyle='arc3, rad=-.9')
    )
```
